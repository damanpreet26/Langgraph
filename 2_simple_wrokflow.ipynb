```python
from langgraph.graph import StateGraph, START, END
from langchain_openai   import ChatOpenAI
from langchain_community.llms import Ollama
from typing import TypedDict
from dotenv import load_dotenv

load_dotenv()
```

```python
class LLMstate(TypedDict):
    question: str
    answer: str
```

```python
## replacing OpenAI key with qwen3:4b hosted on Ollama

# model=ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)
model = Ollama(model="qwen3:4b", temperature=0.7)
```

```python
def llmqa(state: LLMstate)-> LLMstate:
    question = state['question']
    
    # prompt = f"Answer the following question: {question}"
    
    answer = model.invoke(question)
    state['answer'] = answer
    return state
```

```python
graph = StateGraph(LLMstate)

graph.add_node("llmqa", llmqa)

graph.add_edge(START, "llmqa")
graph.add_edge("llmqa", END)

workflow=graph.compile()

workflow.invoke({
    "question": "What is the capital of punjab, india?"
})
```
